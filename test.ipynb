```python
import os
def walk_root_dir(img_root):
    names=[]
    wholePathes=[]
    annoNames=[]
    annoWholePathes=[]
    for dirpath, subdirs, files in os.walk(img_root):
        for x in files:
            if x.endswith(('.png','.jpg', '.jpeg', '.tif')):
                names.append(x)
                wholePathes.append(os.path.join(dirpath, x))
                annoWholePathes.append(os.path.join('./datasets/FIVES/train/Groundtruth/', x))
    return names, wholePathes, annoWholePathes
```

```python
names, wholePathes, annoWholePathes = walk_root_dir('./datasets/FIVES/train/Original/')
```

```python
labels = []
for name in names:
    labels.append(name.split('.')[0][-1])
```
